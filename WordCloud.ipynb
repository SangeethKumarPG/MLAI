```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
text = '''I have to remind myself that some birds aren’t meant to be caged. Their feathers are just too bright. And when they fly away, the part of you that knows it was a sin to lock them up does rejoice. I hope to see my friend and shake his hand. I hope the Pacific is as blue as it has been in my dreams. I hope.
Hope is a good thing, maybe the best of things, and no good thing ever dies. I guess it comes down to a simple choice, really. Get busy living or get busy dying. Hope is a dangerous thing. Hope can drive a man insane.
A whole life blown away in the blink of an eye. Nothing left but all the time in the world to think about it. I remember thinking it would take a man 600 years to tunnel through the wall with it. Andy did it in less than 20.
'''
```

```python
text
```

```python
f_text = text.replace('\n','')
```

```python
f_text
```

```python
nltk.download("stopwords")
nltk.download("punkt")
```

```python
tokens = word_tokenize(f_text)
```

```python
from PIL import Image
image = Image.open("/content/Image 01-11-23 at 8.23 PM.jpeg")
image = image.resize((800,400))
import numpy as np
image_array = np.array(image)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
wordcloud = WordCloud(width=800, height=400, background_color="black").generate("".join(filtered_tokens))
```

```python
word_cloud_array = np.array(wordcloud)
super_imposed_image = np.where(word_cloud_array == 0, image_array, word_cloud_array)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(super_imposed_image, interpolation="bilinear")
plt.axis("off")
plt.show()
```
