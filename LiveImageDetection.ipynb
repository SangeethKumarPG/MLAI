### Importing libraries

```python
import numpy as np
import cv2
import tensorflow as tf
```

loading model

```python
cnn = tf.keras.models.load_model("trained_model_optimised1.h5")
```

```python
class_names = ['Bean',
 'Bitter_Gourd',
 'Bottle_Gourd',
 'Brinjal',
 'Broccoli',
 'Cabbage',
 'Capsicum',
 'Carrot',
 'Cauliflower',
 'Cucumber',
 'Papaya',
 'Potato',
 'Pumpkin',
 'Radish',
 'Tomato']
```

```python
import cv2

# Initialize the video capture
cap = cv2.VideoCapture(0)
image_size = (64, 64)
roi_width = 200  
roi_height = 200  


screen_width = 1920  
screen_height = 1080  

# Calculate the coordinates for frame
x = (screen_width - roi_width) // 2
y = (screen_height - roi_height) // 2

while True:
    ret, frame = cap.read()

    if not ret:
        break

    #region of frame
    roi = frame[y:y+roi_height, x:x+roi_width]
    roi = cv2.resize(roi, image_size)
    
    # Convert the color to RGB
    roi = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)

    # Converting the frame to image array and predicting
    image_array = tf.keras.preprocessing.image.img_to_array(roi)
    input_array = np.array([image_array])
    prediction = cnn.predict(input_array)
    predicted_class_index = np.argmax(prediction)
    predicted_class = class_names[predicted_class_index]

    # Put the predicted class label on the Frame
    cv2.putText(frame, predicted_class, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (100, 200, 255), 2)

    # Drawing a rectangle for frame
    cv2.rectangle(frame, (x, y), (x + roi_width, y + roi_height), (0, 255, 0), 2)

    cv2.imshow("Object Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
